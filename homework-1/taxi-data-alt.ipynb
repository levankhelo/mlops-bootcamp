```python
import pandas as pd

# Read the parquet file
df = pd.read_parquet('yellow_tripdata_2023-01.parquet')

# Display the dataframe
df.head()
```

```python
# Calculate the number of columns
num_columns = len(df.columns)
print(f"Q1: Downloading the data. Columns: {num_columns}")
```

```python
# Compute the duration variable
df['duration'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df.head()
```

```python
# Compute the standard of the duration variable
duration_std = df['duration'].std()
print(f"Q2: computing duration. Standard deviation: {duration_std}")
```

```python
# Filter the dataframe to keep only records with duration between 1 and 60 minutes (including)
filtered_df = df[(df['duration'] >= pd.Timedelta(minutes=1)) & (df['duration'] <= pd.Timedelta(minutes=60))]
filtered_df.head()
```

```python

# Calculate the fraction of records left
fraction_left = len(filtered_df) / len(df)
print(f"Q3: Dropping outliers. fraction left after filter: {fraction_left*100:.0f}% - actual value: {fraction_left}")
```

```python

# Select only the pickup and dropoff location ID columns and recast as a str
filtered_df.loc[:,'PULocationID'] = filtered_df['PULocationID'].astype(str)
filtered_df.loc[:,'DOLocationID'] = filtered_df['DOLocationID'].astype(str)
# Convert dataframe to dictionaries
dicts = filtered_df[['PULocationID', 'DOLocationID']].to_dict(orient='records')
```

```python
from sklearn.feature_extraction import DictVectorizer
# Fit a dictionary vectorizer
dv = DictVectorizer(sparse=False)
feature_matrix = dv.fit_transform(dicts)

print(f"Q4: One-hot encoding. Feature matrix shape: {feature_matrix.shape} - columns: {feature_matrix.shape[1]} rows: {feature_matrix.shape[0]}")
```

```python
# Create the target variable as the duration in minutes
target = filtered_df['duration'].dt.total_seconds() / 60  # Convert duration to minutes
# print dimensions for comparison with feature_matrix
target.shape
```

```python
# print dimensions for comparison with target
feature_matrix.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize the Linear Regression model
lr = LinearRegression()

# Fit the model using the feature matrix and the target variable
lr.fit(feature_matrix, target)

# Train: Predict the target values on the training data
predictions = lr.predict(feature_matrix)

# Calculate the RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(target, predictions))
# Output the RMSE value
print(f"RMSE on training data: {rmse}")
```
